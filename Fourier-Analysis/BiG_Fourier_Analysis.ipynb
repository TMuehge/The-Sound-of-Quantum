```python
# Filter out dead noise
def trimmed(wav, threshold):
    # Absolute value
    wav = np.abs(wav)
    # Point wise boolean mask determination.
    mask = wav > threshold
    return wav[mask]
```

```python
plt.figure()
librosa.display.waveshow(y = samples, sr = sampling_rate)
plt.xlabel('Time (seconds) -->')
plt.ylabel('Amplitude')
plt.show()
```

```python
def fft_components(sound):
    # Find FFT
    fft = np.fft.fft(sound)
    # Find magnitude
    mag = np.abs(np.real(fft))
    # Find frequency
    freq = np.linspace(0,sr, len(mag))
    return mag, freq
```

```python
k_mag, k_freq = fft_components(samples)
```

```python
# Plotting the kick drum
plt.plot(k_freq, k_mag)
plt.show()
```
